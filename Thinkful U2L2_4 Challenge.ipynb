#### Contraceptive Method Choice - What Do You See?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

In order to set up my Jupyter Notebook environment properly, I began by importing numpy, pandas, and matplotlib.  I need to check on this error message about a possible binary incompatibility.  Next I wanted to create a data frame to see the data, with descriptive column names.

```python
df = pd.read_csv('cmc.csv')
df.columns = ["Mother's Age", "Mother's Education", "Father's Education", "# of Children Ever", "Islamic Mother", "Stay at Home", "Husband's Job", "Standard of Living", "Media Exposure", "Contraception"]
print(df.head)
```

This is quite a large data set, so next I want to find my min and max ages for my mothers.

```python
age_mother = df[1:]

print(np.max(age_mother))
print(np.min(age_mother))
```

Next I wanted to separate each age group and find the mean of each column for each age group.  This was done to 
simply get a general idea of the relationships among the values.

```python
df.groupby("Mother's Age").aggregate(np.mean)
```

Now that I can see some relationships among the values, the next visualization I would like to see is a histogram
showing the distribution of the number of women that makes up each age grouping.

```python
# Generate 500 random values.
x = np.random.normal(25, 10, 500)

# Plot them as a histogram.
plt.hist(x) 
plt.show()
```

Next I used set up a boxplot showing the ages of the women.  It appears that some of the other data has also been
included in this plot.

```python
np.random.seed(111)
x = np.random.normal(32, 8)

plt.boxplot(age_mother)
plt.show()
```

A scatter plot would be more effective if the values weren't exactly 1, 2, or 3 for the contraception variable.  It would also be helpful if the size of the "x" would increase in size as more values are placed on top.

```python
plt.scatter(
    x=df["Mother's Age"],
    y=df["Contraception"],
    color='purple',
    marker='x', s=10
)
plt.show()
```

My last visualization is a line plot with all of the values included.  This might be better as subplots.

```python
df.plot(kind='line')
plt.show()
```

Now it is much easier to see the individual values of the subplots.  It is interesting to note that approximately 600 women were using no contraception, a number that is approaching 50%.  I would be interested in seeing if the use of contraception increases with age or if it decreases with age.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df["Mother's Age"], color='red')
plt.ylabel("Age")
plt.title("Mother's Age")

plt.subplot(1, 2, 2)
plt.plot(df["Contraception"], color='green')
plt.ylabel("1=None, 2=Long-Term, 3=Short-Term")
plt.title("Contraception")
plt.show()
```
